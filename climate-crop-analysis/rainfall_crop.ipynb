```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../assets/data/crop_rainfall_sample.csv')
df.head()
```

```python
# scatter + regression
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='annual_rainfall_mm', y='crop_yield_ton_per_ha')
plt.title("Rainfall vs Crop Yield")
plt.xlabel("Annual Rainfall (mm)")
plt.ylabel("Yield (ton/ha)")
plt.show()
```

```python
# compute correlation
corr = df['annual_rainfall_mm'].corr(df['crop_yield_ton_per_ha'])
print(f"Pearson correlation: {corr:.2f}")
```
